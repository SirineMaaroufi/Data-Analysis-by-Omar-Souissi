```python
import pandas as pd
import numpy as np
```

```python
import csv
```

```python
d = pd.read_csv('Cours_BTC2FEUR_-_Feuille_1.csv')
```

```python
d.head()
```

```python
d.describe()
```

```python
d.reindex(columns = ['Date','Variation 24h', 'Prix']).head()
```

```python
d.sort_values(by = 'Prix')
```

```python
s1=d.Prix.apply(lambda x: x.replace("1 BTC = ",""))
d['Prix'] = pd.Series(s1, index = d.index)
d.head()
```

```python
s2=d.Prix.apply(lambda x: x.replace("€",""))
d['Prix'] = pd.Series(s2, index = d.index)
d.head()
```

```python
d[['Prix']] = d[['Prix']].astype(float)
```

```python
d.sort_values(by = 'Prix')
```
